```python
class BPETokenizer:

  def __init__(self, special_tokens: list[str], vocabulary_size: int, lowercase: bool = False):
      self.special_tokens = special_tokens
      self.vocabulary_size = vocabulary_size
      self.lowercase = lowercase
      self.end_of_word = "</w>"
      self.use_unk_token = '[UNK]' in special_tokens
      self.use_bounds_tokens = '[SOS]' in special_tokens and '[EOS]' in special_tokens
      self.vocabulary_size = vocabulary_size
      self.lowercase = lowercase

      self._init_vocabulary()

  def _init_vocabulary(self, id2token = None,token2id = None):
      """Инициализация словая, состоящего из специальных символов и цифр"""
      if id2token is not None and token2id is not None:
          self.id2token = id2token
          self.token2id = token2id
      else:
          special_tokens_list = list(self.special_tokens.values())
          tokens_to_add = special_tokens_list + list(map(str, range(10)))
          self.id2token = {i: token for i, token in enumerate(tokens_to_add)}
          self.token2id = {token: i for i, token in enumerate(tokens_to_add)}
      self.special_token_ids = [self.token2id.get(token, -1) for token in self.special_tokens]

  def get_words(self,pandas_data_frame):
      """Разбивает слова на символы. К последнему символу прибавляем символ конца слова."""

      word_counts = defaultdict(int)

      for line in pandas_data_frame:
                if self.lowercase:
                    line = line.lower().strip()
                else:
                    line = line.strip()
                line = re.sub(r'\d+', '', line)

                for word in line.split():
                    word_counts[word] += 1

      word_counts = dict([(tuple(x[:-1]) + (x[-1] + self.end_of_word,), y) for (x, y) in word_counts.items()])
      return word_counts

  @staticmethod
  def prepare_tokens(word_counts: dict[tuple, int]) -> list[str]:
      """Извлекает и сортирует уникальные токены из словаря слов."""
      all_tokens = set()
      for word_tokens in word_counts.keys():
            all_tokens.update(word_tokens)

        # Сортируем по длине (от самых длинных к самым коротким)
      sorted_tokens = sorted(all_tokens, key=len, reverse=True)
      return sorted_tokens

  @staticmethod
  def get_pair_statistics(word_counts):
      """Считает количество пар символов"""
      pairs = {}
      for word, frequency in word_counts.items():
        for i in range(len(word)-1):
          pair = (word[i], word[i + 1])
          pairs[pair] = pairs.get(pair, 0) + frequency
      return pairs

  @staticmethod
  def merge(word_counts: dict[tuple, int], most_frequent_pair: tuple):
      """most_frequent_pair - пара, которая встречается наиболее часто. Заменяет два символа парой"""
      token1, token2 = most_frequent_pair
      merged_token = token1 + token2
      result_vocab = {}

      for word_tokens, frequency in word_counts.items():
          word_as_string = ' '.join(word_tokens)

          pattern = re.compile(r'(?<!\S)' + re.escape(token1 + ' ' + token2) + r'(?!\S)')
          updated_word_string = pattern.sub(merged_token, word_as_string)

          updated_word_tokens = tuple(updated_word_string.split(' '))
          result_vocab[updated_word_tokens] = frequency

      return result_vocab
  def _update_vocabulary(self, word_counts: dict[tuple, int]):
      """Обновляем словарь, записываем все сочетания символов, а также все уникальные символы"""
      tokens = self.prepare_tokens(word_counts)
      vocab_size = len(self.id2token)


      for i, token in enumerate(tokens):
          self.id2token[i + vocab_size] = token
          self.token2id[token] = i + vocab_size


      symbols = set(''.join(tokens))
      vocab_size, added = len(self.id2token), 0
      for symbol in symbols:
          if symbol not in self.token2id:
            self.id2token[added + vocab_size] = symbol
            self.token2id[symbol] = added + vocab_size
            added += 1

      # Add end of word symbol
      self.id2token[added + vocab_size] = self.end_of_word
      self.token2id[self.end_of_word] = added + vocab_size

  def train(self, pandas_data_frame: str):
        words = self.get_words(pandas_data_frame)

        for i in range(self.vocabulary_size):
            if i % 50 == 0: print(i,self.vocabulary_size)
            pairs = self.get_pair_statistics(words)
            most_frequent_pair = max(pairs, key=pairs.get)
            words = self.merge(words, most_frequent_pair)

        self._update_vocabulary(words)

  def encode(self, text: str, tokenize: bool = False):
      """Преобразует текст в токены"""
      text = text.lower() if self.lowercase else text
      words = [tuple(word) + (self.end_of_word,) for word in text.strip().split()]
      if self.use_bounds_tokens:
          words = [('[SOS]',)] + words + [('[EOS]',)]
      encoded = []
      tokens = []

      for word in words:
          i = 0
          while i < len(word):
              unknown = True
              for j in range(len(word), i, -1):
                  subword = ''.join(word[i:j])
                  if subword in self.token2id:
                      encoded.append(self.token2id[subword])
                      tokens.append(subword)
                      i = j - 1
                      unknown = False
                      break
              i += 1
              if unknown and self.use_unk_token:
                  encoded.append(self.token2id['[UNK]'])
                  tokens.append('[UNK]')
      if tokenize:
          return tokens
      return encoded

  def decode(self, tokens, skip_special_tokens = True) -> str:
      """Преобразует  токены в текст"""
      if skip_special_tokens:
          decoded = ''.join([self.id2token[t] for t in tokens if t not in self.special_token_ids])
      else:
          decoded = ''.join([self.id2token[t] for t in tokens])

      if self.use_bounds_tokens and not skip_special_tokens:

          decoded = decoded.replace('[SOS]', f'[SOS]{self.end_of_word}').replace('[EOS]', f'[EOS]{self.end_of_word}')

      return decoded.replace(self.end_of_word, ' ').strip()

  def tokenize(self, text: str) -> list[str]:
      return self.encode(text, tokenize=True)

  def save(self, path: str):
      """Saves vocabulary state."""
      with open(path, 'wb') as f:
          pickle.dump({'id2token': self.id2token, 'token2id': self.token2id}, f)

  def load(self, path: str):
      """Loads vocabulary state."""
      with open(path, 'rb') as f:
          vocab_mappings = pickle.load(f)

      self._init_vocabulary(vocab_mappings['id2token'], vocab_mappings['token2id'])

  def get_vocab_size(self) -> int:
      """Gets vocabulary size."""
      return len(self.id2token)
```

```python
class TranslationDataset:
    def __init__(self, pairs, src_tokenizer, trgt_tokenizer):
        self.pairs = pairs
        self.src_tokenizer = src_tokenizer
        self.trgt_tokenizer = trgt_tokenizer

    def __len__(self):
        return len(self.pairs)

    def encode(self, sentence, tokenizer):
        return tokenizer.encode(sentence)

    def __getitem__(self, index):
        pair = self.pairs[index]  # Получаем словарь из списка
        src_encoded = self.encode(pair["src"], self.src_tokenizer)
        trgt_encoded = self.encode(pair["trgt"], self.trgt_tokenizer)

        return torch.tensor(src_encoded), torch.tensor(trgt_encoded)
```

```python
def get_sequence_mask(sequence: torch.Tensor, pad_idx: int = 0, mask_future_positions: bool = False,
                      device: str = 'cpu') -> torch.Tensor:
    """Создает маску для последовательностей.

    Аргументы:
        sequence: Тензор с последовательностями.
        pad_idx: Индекс padding-токена (заполнения).
        mask_future_positions: Нужно ли маскировать будущие позиции (для декодера).
        device: Устройство для выполнения операций.

    Возвращаем:
        Тензор с маской.
    """
    batch_size, batch_max_seq_len = sequence.size()
    padding_mask = (sequence == pad_idx).unsqueeze(1).unsqueeze(2)

    if mask_future_positions:
        attention_shape = (batch_max_seq_len, batch_max_seq_len)
        future_positions_mask = torch.triu(torch.ones(attention_shape, device=device), diagonal=1).bool()
        return torch.max(padding_mask, future_positions_mask)

    return padding_mask
```

```python
def collate_function(batch: list):
    """Эта функция преобразует батч последовательностей разной длины в
    padded тензоры одинакового размера. Она подготавливает данные для модели seq2seq,
    создавая encoder_inputs (без служебных токенов), decoder_inputs (без EOS) и decoder_outputs (без SOS).
    Также функция генерирует маски для внимания, скрывающие padding-токены и будущие позиции в декодере.
    """
    encoder_inputs, decoder_inputs, decoder_outputs, sample_indices = [], [], [], []

    for i, (src_tensor, trgt_tensor) in enumerate(batch):
        # Предполагаем, что служебные токены уже добавлены
        encoder_inputs.append(src_tensor[1:-1])  # Без SOS и EOS
        decoder_inputs.append(trgt_tensor[:-1])  # Без EOS
        decoder_outputs.append(trgt_tensor[1:])  # Без SOS
        sample_indices.append(torch.tensor(i, dtype=torch.int))  # Добавляем индекс в батче

    encoder_inputs = pad_sequence(encoder_inputs, batch_first=True)
    decoder_inputs = pad_sequence(decoder_inputs, batch_first=True)
    decoder_outputs = pad_sequence(decoder_outputs, batch_first=True)
    sample_indices = torch.stack(sample_indices)  # Создаем тензор из индексов

    encoder_mask = get_sequence_mask(encoder_inputs)
    decoder_mask = get_sequence_mask(decoder_inputs, mask_future_positions=True)

    return sample_indices, encoder_inputs, decoder_inputs, decoder_outputs, encoder_mask, decoder_mask
```

```python
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_seq_length=100):
        super(PositionalEncoding, self).__init__()

        # Создаем позиционное кодирование для максимальной длины
        pe = torch.zeros(max_seq_length, d_model)
        position = torch.arange(0, max_seq_length, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-torch.log(torch.tensor(10000.0)) / d_model))

        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)


        self.register_buffer('pe', pe.unsqueeze(0))  # [1, max_seq_length, d_model]

    def forward(self, x):
        # x: [batch_size, seq_len, d_model]
        seq_len = x.size(1)

        # Берем только нужную часть позиционного кодирования
        return x + self.pe[:, :seq_len, :]
```

```python
class Encoder(nn.Module):
    def __init__(self, nhead, d_model, dim_feedforward, num_layers, dropout=0.1):
        super(Encoder, self).__init__()
        self.layers = nn.ModuleList([
            EncoderLayer(d_model, nhead, dim_feedforward, dropout)
            for _ in range(num_layers)
        ])
        self._reset_parameters()

    def _reset_parameters(self):
        for p in self.parameters():
            if p.dim() > 1:
                nn.init.xavier_uniform_(p)

    def forward(self, src, src_mask=None, src_key_padding_mask=None):
        output = src
        for layer in self.layers:
            output = layer(output, src_mask=src_mask, src_key_padding_mask=src_key_padding_mask)
        return output

class EncoderLayer(nn.Module):
    def __init__(self, d_model, nhead, dim_feedforward=2048, dropout=0.1):
        super(EncoderLayer, self).__init__()
        self.self_attn = nn.MultiheadAttention(
            d_model, nhead, dropout=dropout, batch_first=True
        )

        self.linear1 = nn.Linear(d_model, dim_feedforward)
        self.dropout = nn.Dropout(dropout)
        self.linear2 = nn.Linear(dim_feedforward, d_model)

        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)
        self.dropout1 = nn.Dropout(dropout)
        self.dropout2 = nn.Dropout(dropout)

        self.activation = nn.ReLU()

    def forward(self, src, src_mask=None, src_key_padding_mask=None):

        src2 = self.norm1(src)
        q = k = v = src2
        src2, attn_weights = self.self_attn(
            q, k, v,
            attn_mask=src_mask,
            key_padding_mask=src_key_padding_mask,
            need_weights=False
        )
        src = src + self.dropout1(src2)


        src2 = self.norm2(src)
        src2 = self.linear2(self.dropout(self.activation(self.linear1(src2))))
        src = src + self.dropout2(src2)

        return src
```

```python
class Decoder(nn.Module):
    def __init__(self, heads: int, embed_size: int, hidden_size: int, num_layers: int):
        super(Decoder, self).__init__()
        self.layers = nn.ModuleList([
            DecoderLayer(heads, embed_size, hidden_size)
            for _ in range(num_layers)
        ])
        self._reset_parameters()

    def _reset_parameters(self):
        for p in self.parameters():
            if p.dim() > 1:
                nn.init.xavier_uniform_(p)

    def forward(self, x, encoder_outputs, tgt_mask=None,
                tgt_key_padding_mask=None, memory_key_padding_mask=None):
        for layer in self.layers:
            x = layer(x, encoder_outputs,
                     tgt_mask=tgt_mask,
                     tgt_key_padding_mask=tgt_key_padding_mask,
                     memory_key_padding_mask=memory_key_padding_mask)
        return x

class DecoderLayer(nn.Module):
    def __init__(self, heads: int, embed_size: int, hidden_size: int, dropout=0.1):
        super(DecoderLayer, self).__init__()

        self.maskedMHA = nn.MultiheadAttention(embed_size, heads, batch_first=True, dropout=dropout)
        self.dropout1 = nn.Dropout(dropout)
        self.norm1 = nn.LayerNorm(embed_size)

        self.MHA = nn.MultiheadAttention(embed_size, heads, batch_first=True, dropout=dropout)
        self.dropout2 = nn.Dropout(dropout)
        self.norm2 = nn.LayerNorm(embed_size)

        self.ffn = nn.Sequential(
            nn.Linear(embed_size, 4 * embed_size),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(4 * embed_size, embed_size)
        )
        self.dropout3 = nn.Dropout(dropout)
        self.norm3 = nn.LayerNorm(embed_size)

    def forward(self, x, encoder_outputs, tgt_mask=None,
                tgt_key_padding_mask=None, memory_key_padding_mask=None):


        self_attention, _ = self.maskedMHA(
            x, x, x,
            attn_mask=tgt_mask,
            key_padding_mask=tgt_key_padding_mask,
            need_weights=False
        )
        x = x + self.dropout1(self_attention)
        x = self.norm1(x)


        cross_attention, _ = self.MHA(
            x, encoder_outputs, encoder_outputs,
            key_padding_mask=memory_key_padding_mask,
            need_weights=False
        )
        x = x + self.dropout2(cross_attention)
        x = self.norm2(x)


        ffn_output = self.ffn(x)
        x = x + self.dropout3(ffn_output)
        x = self.norm3(x)

        return x
```

```python
class OptimusPrime(nn.Module):
    def __init__(self, src_vocab_size: int, tgt_vocab_size: int, d_model: int = 512,
                 num_heads: int = 8, num_layers: int = 6, hidden_size: int = 2048,
                 dropout: float = 0.1):
        super(OptimusPrime, self).__init__()

        self.d_model = d_model
        self.src_vocab_size = src_vocab_size
        self.tgt_vocab_size = tgt_vocab_size


        self.src_embedding = nn.Embedding(src_vocab_size, d_model)
        self.tgt_embedding = nn.Embedding(tgt_vocab_size, d_model)


        self.positional_encoding = PositionalEncoding(d_model)

        self.encoder = Encoder(num_heads, d_model, hidden_size, num_layers)
        self.decoder = Decoder(num_heads, d_model, hidden_size, num_layers)

        self.output_layer = nn.Linear(d_model, tgt_vocab_size)

        self._init_weights()

    def _init_weights(self):
        for name, p in self.named_parameters():
            if p.dim() > 1:
                nn.init.xavier_uniform_(p)
            elif 'bias' in name:
                nn.init.constant_(p, 0)

        nn.init.normal_(self.src_embedding.weight, mean=0, std=self.d_model**-0.5)
        nn.init.normal_(self.tgt_embedding.weight, mean=0, std=self.d_model**-0.5)



    def forward(self, src_tokens, tgt_tokens, encoder_mask=None, decoder_mask=None):
        if encoder_mask is not None and decoder_mask is not None:
            src_key_padding_mask = encoder_mask.squeeze(1).squeeze(1)
            tgt_key_padding_mask = decoder_mask.squeeze(1).squeeze(1)

            seq_len = tgt_tokens.size(1)
            tgt_mask = torch.triu(torch.ones(seq_len, seq_len, device=tgt_tokens.device) * float('-inf'), diagonal=1)
        else:
            src_key_padding_mask = (src_tokens == 0)
            tgt_key_padding_mask = (tgt_tokens == 0)
            seq_len = tgt_tokens.size(1)
            tgt_mask = torch.triu(torch.ones(seq_len, seq_len, device=tgt_tokens.device) * float('-inf'), diagonal=1)

        src_embedded = self.src_embedding(src_tokens) * (self.d_model)**0.5
        src_embedded = self.positional_encoding(src_embedded)


        tgt_embedded = self.tgt_embedding(tgt_tokens) * (self.d_model)**0.5
        tgt_embedded = self.positional_encoding(tgt_embedded)


        memory = self.encoder(
            src_embedded,
            src_key_padding_mask=src_key_padding_mask
        )

        output = self.decoder(
            tgt_embedded,
            memory,
            tgt_mask=tgt_mask,
            tgt_key_padding_mask=tgt_key_padding_mask,
            memory_key_padding_mask=src_key_padding_mask
        )

        logits = self.output_layer(output)
        return logits


    def generate(self, src, max_length=50, sos_token=1, eos_token=2):
        self.eval()
        batch_size = src.size(0)
        device = src.device


        generated = torch.ones(batch_size, 1).fill_(sos_token).long().to(device)

        with torch.no_grad():
            for _ in range(max_length - 1):
                logits = self(src, generated)

                next_token_logits = logits[:, -1, :]
                next_token = next_token_logits.argmax(dim=-1).unsqueeze(1)

                generated = torch.cat([generated, next_token], dim=1)

                if (next_token == eos_token).all():
                    break

        return generated
```

```python
from evaluate import load
import torch
from tqdm import tqdm
import time

bleu_metric = load("bleu")

def validate_bleu_fast(model, valid_loader, device, trgt_tokenizer, max_samples=None):
    model.eval()
    all_predictions = []
    all_references = []

    if max_samples:
        total_batches = min(len(valid_loader), (max_samples + valid_loader.batch_size - 1) // valid_loader.batch_size)
    else:
        total_batches = len(valid_loader)



    start_time = time.time()
    batch_times = []

    with torch.no_grad():
        for batch_idx, batch_data in enumerate(valid_loader):
            if batch_idx > 200: break
            batch_start = time.time()


            sample_indices, src, trgt_input, trgt_output, encoder_mask, decoder_mask = batch_data
            src = src.to(device, non_blocking=True)

            generated = model.generate(src)

            pred_tokens_batch = generated.tolist()
            ref_tokens_batch = trgt_output.tolist()

            for i in range(len(pred_tokens_batch)):
                pred_text = trgt_tokenizer.decode(pred_tokens_batch[i], skip_special_tokens=True)
                ref_text = trgt_tokenizer.decode(ref_tokens_batch[i], skip_special_tokens=True)

                all_predictions.append(pred_text)
                all_references.append([ref_text])

            batch_time = time.time() - batch_start
            batch_times.append(batch_time)


    total_time = time.time() - start_time

    if all_predictions:

        results = bleu_metric.compute(predictions=all_predictions, references=all_references)
        print(f"BLEU Validation completed: {len(all_predictions)} samples, Time: {total_time:.2f}s")
        print(results)
        return results['bleu']
    return 0.0
```

```python

def train():
  num_epochs = 3
  train_losses = []
  valid_losses = []
  bleu_scores = []

  for epoch in range(num_epochs):
      model.train()
      total_train_loss = 0

      pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}", unit="batch")
      start_time = time.time()
      batch_start_time = time.time()

      for batch_idx, batch_data in enumerate(pbar, start=1):
          sample_indices, src, trgt_input, trgt_output, encoder_mask, decoder_mask = batch_data
          src, trgt_input, trgt_output = src.to(device), trgt_input.to(device), trgt_output.to(device)
          optimizer.zero_grad()

          logits = model(src, trgt_input)

          loss = criterion(
              logits.reshape(-1, logits.size(-1)),
              trgt_output.reshape(-1)
          )

          loss.backward()
          optimizer.step()
          total_train_loss += loss.item()

          avg_train_loss = total_train_loss / batch_idx
          batch_time = time.time() - batch_start_time
          batch_start_time = time.time()



          pbar.set_postfix({
              "Train Loss": f"{avg_train_loss:.4f}",
              "Cur Loss": f"{loss.item():.4f}",
              "Batch time": f"{batch_time:.3f}s",
              "ETA": f"{(batch_time)*(len(train_loader)-batch_idx)/60:.1f}m"
          })

      epoch_train_loss = total_train_loss / len(train_loader)
      train_losses.append(epoch_train_loss)

      # Логирование эпохи тренировки
      run["train/epoch_loss"].append(epoch_train_loss)
      run["train/epoch"].append(epoch + 1)

      print(f"Epoch {epoch+1}, Final Train Loss: {epoch_train_loss:.4f}")

      model.eval()
      total_valid_loss = 0

      valid_start_time = time.time()

      with torch.no_grad():
        valid_batch_count = 0
        total_valid_loss = 0

        valid_start_time = time.time()

        for valid_batch_idx, batch_data in enumerate(valid_loader, start=1):
            sample_indices, src, trgt_input, trgt_output, encoder_mask, decoder_mask = batch_data
            src, trgt_input, trgt_output = src.to(device), trgt_input.to(device), trgt_output.to(device)

            logits = model(src, trgt_input)
            loss = criterion(
                logits.reshape(-1, logits.size(-1)),
                trgt_output.reshape(-1)
            )

            if not torch.isnan(loss):
                total_valid_loss += loss.item()
                valid_batch_count += 1
                current_loss = loss.item()

            else:
                current_loss = float('nan')
                continue

            if valid_batch_count > 0:
                avg_valid_loss = total_valid_loss / valid_batch_count
            else:
                avg_valid_loss = float('nan')

            valid_batch_time = time.time() - valid_start_time
            valid_start_time = time.time()

      if valid_batch_count > 0:
          epoch_valid_loss = total_valid_loss / valid_batch_count
      else:
          epoch_valid_loss = float('nan')
          print("⚠️ Все батчи в валидации содержали NaN!")

      valid_losses.append(epoch_valid_loss)

      # Логирование валидационных метрик
      if not torch.isnan(torch.tensor(epoch_valid_loss)):
          run["valid/epoch_loss"].append(epoch_valid_loss)

      print(f"Epoch {epoch+1}, Valid Loss: {epoch_valid_loss:.4f}")

      # Вычисление BLEU score
      if str(device) == "mps" or (hasattr(device, 'type') and device.type == "mps"):
          model_cpu = model.to("cpu")
          bleu_score = validate_bleu_fast(model_cpu, valid_loader, torch.device("cpu"))
          model = model.to(device)
      else:
          bleu_score = validate_bleu_fast(model, valid_loader, device, trgt_tokenizer)

      bleu_scores.append(bleu_score)

      # Логирование BLEU score
      run["valid/bleu_score"].append(bleu_score)

      print(f"Epoch {epoch+1}, BLEU Score: {bleu_score:.4f}")

      print("-" * 60)

  # ===== ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ =====
  print("\n=== ИТОГИ ОБУЧЕНИЯ ===")
  for epoch in range(num_epochs):
      print(f"Epoch {epoch+1}: Train Loss: {train_losses[epoch]:.4f}, "
            f"Valid Loss: {valid_losses[epoch]:.4f}, BLEU: {bleu_scores[epoch]:.4f}")

  # Логирование финальных результатов
  run["final/train_loss"] = train_losses[-1]
  run["final/valid_loss"] = valid_losses[-1] if not torch.isnan(torch.tensor(valid_losses[-1])) else None
  run["final/bleu_score"] = bleu_scores[-1]

  # Завершение сессии Neptune
  run.stop()
```
